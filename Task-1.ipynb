### Data Visualization

```python
import matplotlib.pyplot as plt

plt.hist(df['rating'],bins=10)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplot for Ratings by Skin Tone
plt.figure(figsize=(20, 10))
sns.boxplot(x='skin_tone', y='rating', data=df, palette='pastel')
plt.title('Ratings Distribution by Skin Tone')
plt.xlabel('Skin Tone')
plt.ylabel('Rating')
plt.show()
```

```python
# Pairplot for Numerical Features
numerical_cols = ['rating', 'helpfulness', 'total_feedback_count', 'price_usd']
sns.pairplot(df[numerical_cols])
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()
```

```python
# Time Series Plot for Submission Time

df.set_index('submission_time', inplace=True)

plt.figure(figsize=(12, 8))
df.resample('M').size().plot()
plt.title('Number of Reviews Over Time')
plt.xlabel('Submission Time')
plt.ylabel('Count')
plt.show()
```
